```python
#task one
import numpy as np
arr = np.ones((5,5))
arr[:,1:4] = 0
print(arr)
#task two
np.random.seed(42)
data = np.random.randn(100, 3)
normalized = (data - data.mean(axis=0)) / data.std(axis=0)
print(normalized)
#task three
X = np.random.randn(50, 3)
true_theta = np.array([2.5, -1.2, 3.7])
y = X @ true_theta + np.random.randn(50) * 0.1
theta = np.linalg.inv(X.T @ X) @ X.T @ y

print('estimated coefficients:', theta)
print('true coefficients:', true_theta)
#estimated coefficients: [ 2.497 -1.198  3.702] have values close to the true coefficients [2.5, -1.2, 3.7].
```
